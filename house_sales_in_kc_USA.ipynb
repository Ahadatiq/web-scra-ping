```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn theme
sns.set_theme(style="whitegrid")

# Visualization 1: Price Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=50, color="blue")
plt.title("Distribution of House Prices", fontsize=16)
plt.xlabel("Price ($)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['sqft_living'], y=df['price'], alpha=0.5)
plt.title("Price vs. Square Footage of Living Space", fontsize=16)
plt.xlabel("Living Space (sqft)", fontsize=12)
plt.ylabel("Price ($)", fontsize=12)
plt.show()
```

```python
# Visualization 3: Price vs. Number of Bedrooms
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['bedrooms'], y=df['price'])
plt.title("Price Variation Across Number of Bedrooms", fontsize=16)
plt.xlabel("Number of Bedrooms", fontsize=12)
plt.ylabel("Price ($)", fontsize=12)
plt.yscale('log')  # Use log scale for better visualization
plt.show()
```

```python
# Visualization 4: Geographic Visualization of Prices
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['long'], y=df['lat'], 
                hue=df['price'], palette="viridis", alpha=0.6)
plt.title("Geographic Distribution of House Prices", fontsize=16)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)
plt.show()
```

```python
# Visualization 5: Renovation Effect on Prices
plt.figure(figsize=(10, 6))
df['renovated'] = df['yr_renovated'].apply(lambda x: 'Yes' if x > 0 else 'No')
sns.boxplot(x='renovated', y='price', data=df)
plt.title("Effect of Renovation on House Prices", fontsize=16)
plt.xlabel("Renovated", fontsize=12)
plt.ylabel("Price ($)", fontsize=12)
plt.yscale('log')  # Use log scale for better visualization
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

# Load the dataset
df = pd.read_csv(r"C:\Users\HP\Documents\Kc_housing_dataset_USA.csv")

# Define the feature columns
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", "view",
            "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]
X = df[features]  # Extract the features from the DataFrame
Y = df['price']   # Extract the target variable (price)

# Define the pipeline steps
steps = [
    ('scale', StandardScaler()),
    ('polynomial', PolynomialFeatures(include_bias=False)),
    ('model', LinearRegression())
]

# Create the pipeline
pipe = Pipeline(steps)

# Fit the pipeline to the data
pipe.fit(X, Y)

# Make predictions
ypipe = pipe.predict(X)

# Print the R-squared score of the model
print(pipe.score(X, Y))
```

```python
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Load the dataset
df = pd.read_csv(r"C:\Users\HP\Documents\Kc_housing_dataset_USA.csv")

# Define the feature columns
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", "view",
            "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]

# Extract features and target variable
x = df[features]
y = df['price']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=1)

# Create and train the Ridge regression model
RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train, y_train)

# Make predictions on the test set
yhat = RidgeModel.predict(x_test)

# Calculate and print the R-squared score
print(r2_score(y_test, yhat))
```
